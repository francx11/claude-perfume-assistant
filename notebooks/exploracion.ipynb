# Exploración del Catálogo de Perfumes

DÍA 3: Exploración del CSV antes de implementar el DataLoader.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/fragrantica.csv')
print(f'Dataset cargado: {len(df)} perfumes')
```

## 1. Exploración Básica

```python
# Primeras filas
df.head()
```

```python
# Tipos de datos y valores no nulos por columna
df.info()
```

```python
# Estadísticas de columnas numéricas
df.describe()
```

## 2. Limpieza de Datos

```python
# Valores nulos por columna
print('Valores nulos por columna:')
print(df.isnull().sum())
```

```python
# Duplicados
print(f'Filas duplicadas: {df.duplicated().sum()}')
```

```python
# Valores únicos en columnas categóricas
print('Valores únicos en season:', df['season'].unique())
print('Valores únicos en gender:', df['gender'].unique())
```

## 3. Análisis de Notas

El campo `notes` es un string con valores separados por comas. Hay que separarlo para analizarlo.

```python
# Separar notas y contar las más comunes
all_notes = df['notes'].str.split(',').explode().str.strip()

print(f'Total de notas únicas: {all_notes.nunique()}')
print('\nTop 10 notas más comunes:')
print(all_notes.value_counts().head(10))
```

```python
# Visualización de las notas más comunes
top_notes = all_notes.value_counts().head(10)

plt.figure(figsize=(10, 5))
top_notes.plot(kind='bar', color='steelblue')
plt.title('Top 10 notas más comunes')
plt.xlabel('Nota')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Cuántas notas tiene cada perfume de media
notes_count = df['notes'].str.split(',').apply(len)
print(f'Media de notas por perfume: {notes_count.mean():.1f}')
print(f'Mínimo: {notes_count.min()} | Máximo: {notes_count.max()}')
```

## 4. Análisis de Marcas

```python
# Perfumes por marca
print(f'Total de marcas: {df["brand"].nunique()}')
print('\nPerfumes por marca:')
print(df['brand'].value_counts())
```

```python
# Rating medio por marca
print('Rating medio por marca:')
print(df.groupby('brand')['rating'].mean().sort_values(ascending=False).round(2))
```

## 5. Distribución por Season y Gender

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['season'].value_counts().plot(kind='bar', ax=axes[0], color='coral')
axes[0].set_title('Distribución por Season')
axes[0].set_xlabel('Season')
axes[0].set_ylabel('Cantidad')
axes[0].tick_params(axis='x', rotation=45)

df['gender'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Distribución por Gender')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

## 6. Conclusiones para el DataLoader

Lo que hemos aprendido y que afecta a la implementación:

- El campo `notes` es un **string separado por comas**, no una lista. Hay que hacer `.split(',')` al leerlo.
- Los valores de `season` son: `spring`, `summer`, `fall-winter`, `all-year`
- Los valores de `gender` son: `masculine`, `feminine`, `unisex`
